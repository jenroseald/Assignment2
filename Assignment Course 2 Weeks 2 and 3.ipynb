### Merge cases and vaccinated DataFrames

```python
# Left DataFrame is vaccinated
# Merge on index
```

```python
covid = pd.merge(vaccinated, cases, left_index=True, right_index=True)
```

```python
covid.head()
```

- Merge duplicates columns common to both DFs
- Subset to drop duplicate columns

```python
# View Covid column duplicates
covid.columns
```

Create subsetted DataFrame DF called covid_info

```python
covid_info = covid[['Province/State_x', 'Country/Region_x', 'Date_x', 'Vaccinated', 'First Dose',
       'Second Dose', 'Deaths', 'Cases', 'Recovered',
       'Hospitalised']]
```

```python
# View covid_info
covid_info.head()
```

```python
# Rename column lables to remove _x
covid_info.rename(columns = {'Province/State_x':'Province/State', 'Country/Region_x':'Country/Region', 'Date_x':'Date'}, inplace = True)
```

```python
# View covid_info
covid_info.head()
```

Sense check merged covid_info DataFrame

```python
covid_info.dtypes
```

```python
# Check covid_info for missing data
covid_info_na = covid_info[covid_info.isna().any(axis=1)]
```

### Convert Date column from object to DateTime

Import DateTime module

```python
import datetime
```

```python
covid_info['Date'] =  pd.to_datetime(covid_info['Date'])
```

```python
covid_info.dtypes
```

### Drop unnecessary columns and group by Province / State dose totals

Group by province / state to sum Vaccinated, First Dose and Second Doses totals

```python
covid_summary = covid_info.groupby('Province/State')[['Vaccinated', 'First Dose', 'Second Dose']].sum()
```

```python
covid_summary
```

Include Date column in Groupby()

```python
covid_summary = covid_info.groupby(['Province/State', 'Date'])[['Vaccinated', 'First Dose', 'Second Dose']].sum()
```

```python
covid_summary
```

Remove Date from Groupby as does not add value to analysis

```python
covid_summary = covid_info.groupby('Province/State')[['Vaccinated', 'First Dose', 'Second Dose']].sum()
```

```python
covid_summary
```

Calculate difference between First Dose and Second Dose

```python
# View covid_summary columns
covid_summary.columns
```

```python
# Add columnn into covid_summary DataFrame to hold difference between first and second dose
covid_summary = pd.DataFrame(covid_summary, columns=['Vaccinated', 'First Dose', 'Second Dose', 'Dose Difference'])
```

```python
covid_summary
```

```python
# Calculate numeric difference between First and Second doses
covid_summary['Dose Difference'] = covid_summary['First Dose'] - covid_summary['Second Dose']
```

```python
covid_summary
```

```python
# Sort on Dose Difference
covid_summary.sort_values(by='Dose Difference', ascending=False)
```

Dose difference as percentages

```python
covid_summary['% Difference'] = 100 * covid_summary['Second Dose'] / covid_summary['First Dose']
```

```python
covid_summary
```

```python
# Sort on percentage Difference
covid_summary.sort_values(by='% Difference', ascending=False)
```
